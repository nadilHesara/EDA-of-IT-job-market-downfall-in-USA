```python
import pandas as pd
from functools import reduce
```

```python
file_configs = [
    {
        "file_path": "D:\ML code works\Telecom Churn Predictor\Data\Job postings - hiring demand dataset\Job-Postings-on-Indeed.csv",
        "rename_map": {
            "observation_date": "date",
            "IHLIDXUS": "indeed_job_postings_index_us"
        }
    },
    {
        "file_path": "D:\ML code works\Telecom Churn Predictor\Data\Job postings - hiring demand dataset\Software-Development-Job-Postings-on-Indeed.csv",
        "rename_map": {
            "observation_date": "date",
            "IHLIDXUSTPSOFTDEVE": "indeed_software_development_job_postings_index_us"
        }
    }
]
```

```python
def load_and_prepare_csv(file_path, rename_map, date_col="date"):
    df = pd.read_csv(file_path)
    df = df.rename(columns=rename_map)
    df[date_col] = pd.to_datetime(df[date_col])
    return df
```

```python
dfs = [load_and_prepare_csv(cfg["file_path"], cfg["rename_map"]) for cfg in file_configs]

merged_df = reduce(lambda left, right: pd.merge(left, right, on="date", how="inner"), dfs)

merged_df.to_csv("Job postings data collection.csv", index=False)
print(merged_df.head())
```
