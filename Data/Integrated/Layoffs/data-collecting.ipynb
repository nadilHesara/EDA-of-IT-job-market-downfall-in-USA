```python
from pathlib import Path

import pandas as pd
```

```python
data_path_candidates = [
    Path("Data/raw/layoffs.csv"),
    Path("../../raw/layoffs.csv"),
]

for data_path in data_path_candidates:
    if data_path.exists():
        break
else:
    raise FileNotFoundError("Could not find layoffs.csv from the current notebook working directory.")

data = pd.read_csv(data_path)
data.head()
```

```python
data.country.unique()
```

```python
USA_data = data[data["country"].str.rstrip(".") == "United States"]
USA_data.head()
```

```python
output_path_candidates = [
    Path("Data/Integrated/Layoffs/Layoffs_USA.csv"),
    Path("Layoffs_USA.csv"),
]

for output_path in output_path_candidates:
    if output_path.parent.exists():
        break

USA_data.to_csv(output_path, index=False)
```
