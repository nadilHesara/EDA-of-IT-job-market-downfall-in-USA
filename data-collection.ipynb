```python
import pandas as pd


excel_files = {
    'hires-level': 'Hires-Level',
    'hires-rate': 'Hires-Rate',
    'job-opening-level': 'Job-Opening-Level',
    'job-opening-rate': 'Job-Opening-Rate',
    'job-openings-profbusiness-level': 'Job-Openings-ProfBusiness-Level',
    'layoffs-discharges-level': 'Layoffs-Discharges-Level',
    'layoffs-discharges-rate': 'Layoffs-Discharges-Rate',
    'total-separations-rate': 'Total-Separations-Rate'
}

dataframes = {}
for file_key, column_name in excel_files.items():
    file_path = f'Data/raw/{file_key}.xlsx'
    df = pd.read_excel(
        file_path,
        skiprows=13,
        engine='openpyxl',
        usecols='B:D'
    )
    df = df.rename(columns={'Value': column_name})
    dataframes[file_key] = df
    print(f"Loaded {file_key}.xlsx - Shape: {df.shape}")
```

```python
for name, df in dataframes.items():
    print(f"\n{name}:")
    print(df.head())
```

```python
print("Missing values in each dataframe:")
for name, df in dataframes.items():
    missing = df.isnull().sum()
    print(f"\n{name}:")
    print(missing)
```

```python
df_combined = dataframes['hires-level'].copy()

for file_key in list(excel_files.keys())[1:]:
    df_combined = pd.merge(
        df_combined,
        dataframes[file_key],
        on=['Period', 'Year'],
        how='outer'  
    )

df_combined = df_combined.sort_values(['Year', 'Period']).reset_index(drop=True)

print(f"Combined dataset shape: {df_combined.shape}")
print(f"\nMissing values in combined dataset:")
print(df_combined.isnull().sum())
df_combined.head(10)
```

```python
print("Combined Dataset Info:")
print(df_combined.info())
print("\nSummary Statistics:")
print(df_combined.describe())
```

```python
df_combined.to_csv('combined_job_data.csv', index=False)
```
